```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

sentences = [
    "I love learning Python, it's so much fun!",
    "This assignment is taking forever and I'm exhausted.",
    "The weather today is absolutely beautiful.",
    "I'm not sure if I understand this concept yet.",
    "My code finally worked after hours of debugging!",
    "I'm frustrated that Git keeps confusing me.",
    "Cairo is a wonderful city to live in.",
    "I can't wait to start my graduate studies.",
    "Debugging errors is a painful process.",
    "I'm proud of the progress I've made this week."
]

for sentence in sentences:
    result = classifier(sentence)[0]
    print(f"{sentence}\n → {result['label']} (score: {result['score']:.2f})\n")
```

```python
from huggingface_hub import InferenceClient
from dotenv import load_dotenv
import os

load_dotenv()
HF_TOKEN = os.environ["HF_TOKEN"]

client = InferenceClient(
    provider="hf-inference",
    api_key=HF_TOKEN,
)

result = client.text_classification(
    "I am really enjoying this course!",
    model="cardiffnlp/twitter-roberta-base-sentiment-latest",
)

print(result)
```
